```python
# Imports :

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Database :

data = pd.read_csv('DATABASE.txt', sep=";", header=None, na_values=['?'])
data
```

```python
# On sÃ©lectionne 3 ans 2007 Ã  2009

data = data.rename(columns=data.iloc[0]).drop(data.index[0])

data["Time_index2"] = data["Date"] + " " + data["Time"]

data["timestamp"] = pd.to_datetime(data['Time_index2'], format='%d/%m/%Y %H:%M:%S')
data.set_index(['timestamp'],inplace=True)



data["Global_active_power"] = data["Global_active_power"].astype(float)

data.head()

df = pd.DataFrame(data, columns=['Global_active_power'])

df = df.rename(columns={'Global_active_power': 'Y'})

mask = (df.index > '2007-1-1') & (df.index <= '2009-12-31')

data = df.loc[mask]

# On retire les valeurs manquantes
data.dropna(inplace = True)

data
```

```python
# Resample :

Mins = data['Y'].resample(rule = 'min').mean()
Hours = data['Y'].resample(rule = 'H').mean()
Days = data['Y'].resample(rule = 'D').mean()
Month = data['Y'].resample(rule = 'M').mean()

Days.to_csv(r'C:\Users\bapti\Documents\Equancy\Machine learning\Comparaison Biblio\Days.csv', index = True)

Days.plot()
```

```python
""""
# SMA DATA : (Simple Mean Average)

# Mins :

data['Mins'] = data['Y'].rolling(window = 60).mean()

# Hours :

data['Hours'] = data['Mins'].rolling(window = 60).mean()

# Days :

data['Days'] = data['Hours'].rolling(window = 24).mean()

data[200:1000].plot()

""""
```
